```python
# Mathematical Calculator

import math
import tkinter as tk
import sympy as sym
import numpy as np
x=sym.Symbol('x')

window=tk.Tk()
window.title('Mathematical Calculator')

# Mathematical Functions

def open_new_window1():
    Functions=tk.Toplevel(master=window)
    Functions.title('Mathematical Functions')

    # Square Root

    def openwindow1():
        Square_root=tk.Toplevel(master=Functions)
        Square_root.title('Square Root')


        fsr=tk.Frame(master=Square_root,bg='violet',width='50')
        fsr.pack()


        l1sr=tk.Label(master=fsr,text='Enter The Number Whose Square root is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1sr.pack()
        esr=tk.Entry(master=fsr,bg='yellow',fg='black',relief=tk.RIDGE)
        esr.pack()


        def SquareRoot():
            l2sr['text']=round(math.sqrt(float(esr.get())),4)

        bsr=tk.Button(master=fsr,bg='lime',fg='black',command=SquareRoot,text='Square Root',relief=tk.RAISED,width='20')
        l2sr=tk.Label(master=fsr,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bsr.pack()
        l2sr.pack()

    b1=tk.Button(master=Functions,text='Square Root',command=openwindow1,bg='aqua',fg='black',width=15)
    b1.grid(row=1,column=0,pady=5,sticky='n')



    # Square of a Number

    def openwindow2():
        Square=tk.Toplevel(master=Functions)
        Square.title('Square')

        fs=tk.Frame(master=Square,bg='violet',width='50')
        fs.pack()


        l1s=tk.Label(master=fs,text='Enter The Number Whose Square is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1s.pack()
        es=tk.Entry(master=fs,bg='yellow',fg='black',relief=tk.RIDGE)
        es.pack()


        def square():
            l2s['text']=round((float(es.get())**2),4)

        bs=tk.Button(master=fs,bg='lime',fg='black',command=square,text='Square',relief=tk.RAISED,width='15')
        l2s=tk.Label(master=fs,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bs.pack()
        l2s.pack()

    b2=tk.Button(master=Functions,text='Square',command=openwindow2,bg='lime',fg='black',width=15)
    b2.grid(row=2,column=0,pady=5,sticky='n')



    # Cube Root

    def openwindow3():
        CubeRoot=tk.Toplevel(master=Functions)
        CubeRoot.title('Cube Root')

        fcr=tk.Frame(master=CubeRoot,bg='violet',width='50')
        fcr.pack()


        l1cr=tk.Label(master=fcr,text='Enter The Number Whose Cube Root is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1cr.pack()
        e1cr=tk.Entry(master=fcr,bg='yellow',fg='black',relief=tk.RIDGE)
        e1cr.pack()


        def CubeRoot():
            l3cr['text']=round((float(e1cr.get()))**(1/3),4)

        bcr=tk.Button(master=fcr,bg='lime',fg='black',command=CubeRoot,text='Cube Root',relief=tk.RAISED,width='15')
        l3cr=tk.Label(master=fcr,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bcr.pack()
        l3cr.pack()

    b3=tk.Button(master=Functions,text='Cube Root',command=openwindow3,bg='aqua',fg='black',width=15)
    b3.grid(row=3,column=0,pady=5,sticky='n')



    # Cube

    def openwindow4():
        Cube=tk.Toplevel(master=Functions)
        Cube.title('Cube')


        fc=tk.Frame(master=Cube,bg='violet',width='50')
        fc.pack()


        l1c=tk.Label(master=fc,text='Enter The Number Whose Cube is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1c.pack()
        e1c=tk.Entry(master=fc,bg='yellow',fg='black',relief=tk.RIDGE)
        e1c.pack()


        def Cube():
            l3c['text']=round((float(e1c.get()))**(3),4)

        bc=tk.Button(master=fc,bg='lime',fg='black',command=Cube,text='Cube',relief=tk.RAISED,width='15')
        l3c=tk.Label(master=fc,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bc.pack()
        l3c.pack()

    b4=tk.Button(master=Functions,text='Cube',command=openwindow4,bg='lime',fg='black',width=15)
    b4.grid(row=4,column=0,pady=5,sticky='n')



    # nth root

    def openwindow5():
        nthRoot=tk.Toplevel(master=Functions)
        nthRoot.title('nthRoot')

        fnr=tk.Frame(master=nthRoot,bg='violet',width='50')
        fnr.pack()


        l1nr=tk.Label(master=fnr,text='Enter The Number Whose nth Root is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1nr.pack()
        e1nr=tk.Entry(master=fnr,bg='yellow',fg='black',relief=tk.RIDGE)
        e1nr.pack()
        l2nr=tk.Label(master=fnr,text='Enter the nth Root Power',bg='pink',fg='black',width='25',relief=tk.GROOVE)
        l2nr.pack()
        e2nr=tk.Entry(master=fnr,bg='yellow',fg='black',relief=tk.RIDGE)
        e2nr.pack()

        def nthroot():
            l3nr['text']=round((float(e1nr.get()))**(1/float(e2nr.get())),4)

        bnr=tk.Button(master=fnr,bg='lime',fg='black',command=nthroot,text='nth root',relief=tk.RAISED,width='15')
        l3nr=tk.Label(master=fnr,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bnr.pack()
        l3nr.pack()

    b5=tk.Button(master=Functions,text='nth Root',command=openwindow5,bg='aqua',fg='black',width=15)
    b5.grid(row=5,column=0,pady=5,sticky='n')



    # nth power

    def openwindow6():
        nthPower=tk.Toplevel(master=Functions)
        nthPower.title('nthPower')


        fnp=tk.Frame(master=nthPower,bg='violet',width='50')
        fnp.pack()


        l1np=tk.Label(master=fnp,text='Enter The Number Whose nth Power is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1np.pack()
        e1np=tk.Entry(master=fnp,bg='yellow',fg='black',relief=tk.RIDGE)
        e1np.pack()
        l2np=tk.Label(master=fnp,text='Enter the nth Power',bg='pink',fg='black',width='25',relief=tk.GROOVE)
        l2np.pack()
        e2np=tk.Entry(master=fnp,bg='yellow',fg='black',relief=tk.RIDGE)
        e2np.pack()

        def nthpower():
            l3np['text']=round((float(e1np.get()))**(float(e2np.get())),4)

        bnp=tk.Button(master=fnp,bg='lime',fg='black',command=nthpower,text='nth power',relief=tk.RAISED,width='15')
        l3np=tk.Label(master=fnp,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bnp.pack()
        l3np.pack()

    b6=tk.Button(master=Functions,text='nth Power',command=openwindow6,bg='lime',fg='black',width=15)
    b6.grid(row=6,column=0,pady=5,sticky='n')



    # Factorial

    def openwindow7():
        Factorial=tk.Toplevel(master=Functions)
        Factorial.title('Factorial')

        ff=tk.Frame(master=Factorial,bg='violet',width='50')
        ff.pack()


        l1f=tk.Label(master=ff,text='Enter The Number Whose Factorial is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1f.pack()
        ef=tk.Entry(master=ff,bg='yellow',fg='black',relief=tk.RIDGE)
        ef.pack()


        def Factorial():
            l2f['text']=round(math.factorial(float(ef.get())),4)

        bf=tk.Button(master=ff,bg='lime',fg='black',command=Factorial,text='Factorial',relief=tk.RAISED,width='15')
        l2f=tk.Label(master=ff,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bf.pack()
        l2f.pack()

    b7=tk.Button(master=Functions,text='Factorial',command=openwindow7,bg='aqua',fg='black',width=15)
    b7.grid(row=7,column=0,pady=5,sticky='n')



    # Base-10 Logarithm

    def openwindow8():
        log10=tk.Toplevel(master=Functions)
        log10.title('Base-10 Logarithm')

        fl=tk.Frame(master=log10,bg='violet',width='50')
        fl.pack()


        l1l=tk.Label(master=fl,text='Enter The Number Whose Base-10 Logarithm is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1l.pack()
        el=tk.Entry(master=fl,bg='yellow',fg='black',relief=tk.RIDGE)
        el.pack()


        def Log10():
            l2l['text']=round(math.log10(float(el.get())),4)

        bl=tk.Button(master=fl,bg='lime',fg='black',command=Log10,text='Base-10 Logarithm',relief=tk.RAISED,width='20')
        l2l=tk.Label(master=fl,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bl.pack()
        l2l.pack()

    b8=tk.Button(master=Functions,text='Base-10 Logarithm',command=openwindow8,bg='lime',fg='black',width=15)
    b8.grid(row=8,column=0,pady=5,sticky='n')

    # Natural Logarithm

    def openwindow9():
        ln=tk.Toplevel(master=Functions)
        ln.title('Natural Logarithm')

        fln=tk.Frame(master=ln,bg='violet',width='50')
        fln.pack()


        l1ln=tk.Label(master=fln,text='Enter The Number Whose Natural Logarithm is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1ln.pack()
        eln=tk.Entry(master=fln,bg='yellow',fg='black',relief=tk.RIDGE)
        eln.pack()


        def Ln():
            l2ln['text']=round(math.log(float(eln.get())),4)

        bln=tk.Button(master=fln,bg='lime',fg='black',command=Ln,text='Natural Logarithm',relief=tk.RAISED,width='20')
        l2ln=tk.Label(master=fln,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bln.pack()
        l2ln.pack()

    b9=tk.Button(master=Functions,text='Natural Logarithm',command=openwindow9,bg='Aqua',fg='black',width=15)
    b9.grid(row=9,column=0,pady=5,sticky='n')



    # Integration

    def openwindow10():
        i=tk.Toplevel(master=Functions)
        i.title('Integration')

        fi=tk.Frame(master=i,bg='violet',width='50')
        fi.pack()

        l1i=tk.Label(master=fi,text='Enter a Equation',bg='pink',fg='black',relief=tk.GROOVE,width='20')
        l1i.pack()
        e1i=tk.Entry(master=fi,bg='yellow',fg='black',relief=tk.RIDGE)
        e1i.pack()

        l2i=tk.Label(master=fi,text='Enter Lower Limit',bg='pink',fg='black',relief=tk.GROOVE,width='20')
        l2i.pack()
        e2i=tk.Entry(master=fi,bg='yellow',fg='black',relief=tk.RIDGE)
        e2i.pack()

        l3i=tk.Label(master=fi,text='Enter Upper Limit',bg='pink',fg='black',relief=tk.GROOVE,width='20')
        l3i.pack()
        e3i=tk.Entry(master=fi,bg='yellow',fg='black',relief=tk.RIDGE)
        e3i.pack()


        def integration():
            l4i['text']=sym.integrate(e1i.get(),(x,float(e2i.get()),float(e3i.get())))


        bi=tk.Button(master=fi,bg='lime',fg='black',command=integration,text='Integration',relief=tk.RAISED,width='15')
        l4i=tk.Label(master=fi,bg='aqua',fg='black',relief=tk.SUNKEN,width='15')

        bi.pack()
        l4i.pack()

    b10=tk.Button(master=Functions,text='Integration',command=openwindow10,bg='lime',fg='black',width=15)
    b10.grid(row=10,column=0,pady=5,sticky='n')
   
    # Differentiation
   
    def openwindow11():
        Differentiation=tk.Toplevel(master=Functions)
        Differentiation.title('Differentiation')

        fd=tk.Frame(master=Differentiation,bg='violet',width='50')
        fd.pack()


        l1d=tk.Label(master=fd,text='Enter The Equation Whose Derivative is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1d.pack()
        e1d=tk.Entry(master=fd,bg='yellow',fg='black',relief=tk.RIDGE)
        e1d.pack()


        def Differentiation():
            l2d['text']=sym.Derivative(e1d.get(),x).doit()

        bd=tk.Button(master=fd,bg='lime',fg='black',command=Differentiation,text='Differentiation',relief=tk.RAISED,width='20')
        l2d=tk.Label(master=fd,bg='aqua',fg='black',relief=tk.SUNKEN,width='10')

        bd.pack()
        l2d.pack()

    b11=tk.Button(master=Functions,text='Differentiation',command=openwindow11,bg='Aqua',fg='black',width=15)
    b11.grid(row=11,column=0,pady=5,sticky='n')


    # Sin Inverse
   
    def openwindow12():
        Sin_Inverse=tk.Toplevel(master=Functions)
        Sin_Inverse.title('Sin Inverse')

        fsi=tk.Frame(master=Sin_Inverse,bg='violet',width='50')
        fsi.pack()


        l1si=tk.Label(master=fsi,text='Enter The Number Whose Sin Inverse is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1si.pack()
        e1si=tk.Entry(master=fsi,bg='yellow',fg='black',relief=tk.RIDGE)
        e1si.pack()


        def Sin_Inverse():
            l3si['text']=math.degrees(np.arcsin(float(e1si.get())))

        bsi=tk.Button(master=fsi,bg='lime',fg='black',command=Sin_Inverse,text='Sin Inverse',relief=tk.RAISED,width='20')
        l2si=tk.Label(master=fsi,text='Sin Inverse Value in Degrees',bg='aqua',fg='black',relief=tk.SUNKEN,width='25')
        l3si=tk.Label(master=fsi,bg='aqua',fg='black',relief=tk.SUNKEN,width='20')

        bsi.pack()
        l2si.pack()
        l3si.pack()

    b12=tk.Button(master=Functions,text='Sin Inverse',command=openwindow12,bg='lime',fg='black',width=15)
    b12.grid(row=12,column=0,pady=5,sticky='n')

    

    #Cos inverse
   
    def openwindow13():
        cos_Inverse=tk.Toplevel(master=Functions)
        cos_Inverse.title('Cos Inverse')

        fc=tk.Frame(master=cos_Inverse,bg='violet',width='50')
        fc.pack()


        l1c=tk.Label(master=fc,text='Enter The Number Whose Cos Inverse is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1c.pack()
        e1c=tk.Entry(master=fc,bg='yellow',fg='black',relief=tk.RIDGE)
        e1c.pack()


        def cos_Inverse():
            l3c['text']=math.degrees(np.arccos(float(e1c.get())))

        bc=tk.Button(master=fc,bg='lime',fg='black',command=cos_Inverse,text='Cos Inverse',relief=tk.RAISED,width='20')
        l2c=tk.Label(master=fc,text='Cos Inverse Value in Degrees',bg='aqua',fg='black',relief=tk.SUNKEN,width='25')
        l3c=tk.Label(master=fc,bg='aqua',fg='black',relief=tk.SUNKEN,width='20')

        bc.pack()
        l2c.pack()
        l3c.pack()
    b13=tk.Button(master=Functions,text='Cos Inverse',command=openwindow13,bg='aqua',fg='black',width=15)
    b13.grid(row=13,column=0,pady=5,sticky='n')

    
    #Tan Inverse
   
    def openwindow14():
        tan_Inverse=tk.Toplevel(master=Functions)
        tan_Inverse.title('Tan Inverse')

        ft=tk.Frame(master=tan_Inverse,bg='violet',width='50')
        ft.pack()


        l1t=tk.Label(master=ft,text='Enter The Number Whose Tan Inverse is Required',bg='pink',fg='black',width='50',relief=tk.GROOVE)
        l1t.pack()
        e1t=tk.Entry(master=ft,bg='yellow',fg='black',relief=tk.RIDGE)
        e1t.pack()


        def tan_Inverse():
            l3t['text']=math.degrees(np.arctan(float(e1t.get())))

        bt=tk.Button(master=ft,bg='lime',fg='black',command=tan_Inverse,text='Tan Inverse',relief=tk.RAISED,width='20')
        l2t=tk.Label(master=ft,text='Tan Inverse Value in Degrees',bg='aqua',fg='black',relief=tk.SUNKEN,width='25')
        l3t=tk.Label(master=ft,bg='aqua',fg='black',relief=tk.SUNKEN,width='20')

        bt.pack()
        l2t.pack()
        l3t.pack()

    b14=tk.Button(master=Functions,text='Tan Inverse',command=openwindow14,bg='lime',fg='black',width=15)
    b14.grid(row=14,column=0,pady=5,sticky='n')
    
b1=tk.Button(master=window,text='Mathematical Functions',command=open_new_window1,bg='lime',fg='black',width=20)
b1.grid(row=1,column=0,pady=5,sticky='n')

window.mainloop()
```
